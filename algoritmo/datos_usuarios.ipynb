```python
from faker import Faker
import random
import pandas as pd
import unidecode
from randimage import get_random_image, show_array

fake = Faker("es_ES")
Faker.seed(123)
random.seed(123) 
fixed_digits = 6

idcode, name, age, sex, tel, email, pref_sex, pref_fumador, pref_conv = [[] for k in range(0,9)] 

for row in range(0,10000):
    # Definir sexo dentro del bucle
    sexo = random.choice(["Mujer", "Hombre"])
    sexo2 = random.choice(["Mujeres", "Hombres", "Indiferente"])
    fum = random.choice(["No fumador", "Indiferente"])
    hab = random.choice(["Muy hablador", "Algo hablador", "Poco hablador"])
    
    #si es aleatoriamente mujer:
    if sexo=="Mujer":
        nombre = str(fake.first_name_female()) + ' ' + str(fake.last_name())
        name.append(nombre)
        sex.append("Mujer")
        idcode.append(fake.unique.random_int(min=111111, max=999999))
        age.append(random.randint(17,85))
        tel.append(fake.phone_number().replace(" ", ""))
        email.append(str(unidecode.unidecode(nombre).lower().replace(" ", "")) + "@" + str(fake.free_email_domain()))
        pref_sex.append(sexo2)
        pref_fumador.append(fum)
        pref_conv.append(hab)
        
    #si es aleatoriamente hombre:
    else:
        nombre = str(fake.first_name_male()) + ' ' + str(fake.last_name())
        name.append(nombre)
        sex.append("Hombre")
        idcode.append(str(fake.unique.random_int(min=111111, max=999999)))
        age.append(random.randint(17,85))
        tel.append(fake.phone_number().replace(" ", ""))
        email.append(str(unidecode.unidecode(nombre).lower().replace(" ", "")) + "@" + str(fake.free_email_domain()))
        pref_sex.append(sexo2)
        pref_fumador.append(fum)
        pref_conv.append(hab)
    
d = {"id_usuario":idcode, "nombre_usuario":name, "edad_usuario":age, "sexo_usuario": sex, "telf_usuario":tel, 
     "correo_usuario": email, "pref_sexo": pref_sex, "pref_fumador": pref_fumador, "pref_conv": pref_conv} 
df = pd.DataFrame(d)
df.head()
```

```python
df['telf_usuario'] = df['telf_usuario'].str.replace(r'\+34[1234589]', r'+346', regex=True)
df
```

```python
import numpy as np

# Lista de posibles hobbies y géneros de música
hobbies = ["practicar deporte", "leer", "ver películas", "tocar un instrumento musical", "viajar", "cocinar", "hacer senderismo", "pintar", "ir al cine", "salir a correr"]
generos_musicales = ["rock", "pop", "jazz", "electrónica", "hip-hop", "clásica", "rock-pop", "indie", "reggaeton"]

municipios = ["Valencia, Spain", "Catarroja, Spain", "Torrent, Spain", "Alboraya, Spain", "Moncada, Spain", "Manises, Spain",
          "Tavernes Blanques, Spain", "Meliana, Spain", "Burjasot, Spain", "Paterna, Spain", "Mislata, Spain",
          "Alaquàs, Spain", "Xirivella, Spain", "Paiporta, Spain", "Alfafar, Spain", "Benetúser, Spain", "Sedaví, Spain",
          "Aldaya, Spain", "Quart de Poblet, Spain"]


# Función para generar descripciones
def generar_descripcion(nombre, edad):
    hobby = random.choice(hobbies)
    genero_musical = random.choice(generos_musicales)
    descripcion = f"¡Hola, soy {nombre.split()[0]}! Tengo {edad} años y me gusta {hobby}. Suelo escuchar música {genero_musical}."
    return descripcion

# Agregar descripciones a DataFrame
df['descripcion'] = [generar_descripcion(nombre, edad) for nombre, edad in zip(df['nombre_usuario'], df['edad_usuario'])]


#Que el 30% de los usuarios tengan como municipio "Valencia, Spain"
df['municipio'] = [random.choice(municipios) for i in range(len(df))]
num30 = int(len(df) * 0.3)
indices = random.sample(range(len(df)), num30)
for ind in indices:
    df.at[ind, 'municipio'] = "Valencia, Spain"

df
```

```python
df.to_csv("usuarios.csv",index_label=None, encoding="utf-8")
```
