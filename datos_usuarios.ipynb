```python
from faker import Faker
import random
import pandas as pd
import unidecode
from randimage import get_random_image, show_array

fake = Faker("es_ES")
Faker.seed(123)
random.seed(123) 
fixed_digits = 6

idcode, name, age, sex, tel, email = [[] for k in range(0,6)] 

for row in range(0,10000):
    # Definir sexo dentro del bucle
    sexo = random.choice(["Mujer", "Hombre"])
    
    #si es aleatoriamente mujer:
    if sexo=="Mujer":
        nombre = str(fake.first_name_female()) + ' ' + str(fake.last_name())
        name.append(nombre)
        sex.append("Mujer")
        idcode.append(fake.unique.random_int(min=111111, max=999999))
        age.append(random.randint(17,85))
        tel.append(fake.phone_number().replace(" ", ""))
        email.append(str(unidecode.unidecode(nombre).lower().replace(" ", "")) + "@" + str(fake.free_email_domain()))
        
    #si es aleatoriamente hombre:
    else:
        nombre = str(fake.first_name_male()) + ' ' + str(fake.last_name())
        name.append(nombre)
        sex.append("Hombre")
        idcode.append(str(fake.unique.random_int(min=111111, max=999999)))
        age.append(random.randint(17,85))
        tel.append(fake.phone_number().replace(" ", ""))
        email.append(str(unidecode.unidecode(nombre).lower().replace(" ", "")) + "@" + str(fake.free_email_domain()))
    
d = {"ID":idcode, "Nombre":name, "Edad":age, "Sexo": sex, "Telefono":tel, "Email": email} 
df = pd.DataFrame(d)
df.head()
```

```python
import numpy as np

# Lista de posibles hobbies y géneros de música
hobbies = ["practicar deportes", "leer", "ver películas", "tocar un instrumento musical", "viajar", "cocinar", "hacer senderismo", "pintar"]
generos_musicales = ["rock", "pop", "jazz", "electrónica", "hip-hop", "clásica", "rock-pop", "indie", "reggaeton"]
fumar = ["No me importa viajar con un fumador", "Prefiero no viajar con un fumador"]


# Función para generar descripciones
def generar_descripcion(nombre, edad):
    hobby = random.choice(hobbies)
    genero_musical = random.choice(generos_musicales)
    fum = random.choice(fumar)
    descripcion = f"¡Hola, soy {nombre}! Tengo {edad} años y me gusta {hobby}. Suelo escuchar música {genero_musical}. {fum}"
    return descripcion

# Agregar descripciones a DataFrame
df['Descripcion'] = [generar_descripcion(nombre, edad) for nombre, edad in zip(df['Nombre'], df['Edad'])]

df
```

```python
df.to_csv("usuarios.csv",index_label=None)
```
