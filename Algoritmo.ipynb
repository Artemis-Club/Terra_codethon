```python
def match(df, fila):
    dia = df.at[fila, 'Dia']
    match_dia = df[df['Dia'] == dia]
    
    coche = df.at[fila,'Disponibilidad_coche']
    print(coche)
    if coche == "Pasajero":
        match_dia = match_dia[~(match_dia['Disponibilidad_coche'] == "Pasajero")]
    return match_dia

fila_de_prueba = 1
filas_coincidentes = match(trayectos, fila_de_prueba)
print(filas_coincidentes)
```
